```
import numpy as np
import matplotlib.pyplot as plt
re= 389  #re=1095 
h=0.0052 #en mettre 
s=0.0049 #en mettre 
musurrho=1.4553e-5 #viscosité cinématique = mu/rho
y=np.linspace(0.0049,0.0101,1000) # a lentrée y est dans [s,s+h]
u=6*((re/s)*musurrho)*((y-s)/h)*(1-((y-s)/h)) # la formule de poiseuille a l'entree 
fig = plt.figure(1, figsize=(7,6))
plt.plot(u,y,label="U_int")
plt.xlabel("Flow velocity u [m/s]")
plt.ylabel("Channel width y [m]")
plt.legend()
```
